### DAY 1 Web Summary with Frontier Models

This is my first project working with frontier LLMs models. The goal is to create a web browser, where I fetch a URL to an models and it will response a summary. Based on the prompts (system and user prompt), the models can response in different texts with different tones.

````python
# Import necessary libraries
import os
import requests
from bs4 import BeautifulSoup
from IPython.display import display, Markdown
from openai import OpenAI
````

````python
# A simple greeting to model
open_ai_ollama = OpenAI(base_url="http://localhost:11434/v1", api_key="ollama")

my_message = [
    {
        "role": "user",
        "content": "Hello, this is my very first call to an "
    }
]

gemma2 = open_ai_ollama.chat.completions.create(
    model="gemma2:2b",
    messages=my_message
)
print(gemma2.choices[0].message.content)
````

### A simple web fetching function

````python
# A class to represent a Webpage
# Some websites need you to use proper headers when fetching them:
headers = {
 "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/117.0.0.0 Safari/537.36"
}

class Website:

    def __init__(self, url):
        """
        Create this Website object from the given url using the BeautifulSoup library
        """
        self.url = url
        response = requests.get(url, headers=headers)
        soup = BeautifulSoup(response.content, 'html.parser')
        self.title = soup.title.string if soup.title else "No title found"
        for irrelevant in soup.body(["script", "style", "img", "input"]):
            irrelevant.decompose()
        self.text = soup.body.get_text(separator="\n", strip=True)

#NOTE: The credit belongs to Ed Donner from Udemy course
````

````python
# Lets try with actual url
pasagius_url = "https://memart.vn/tin-tuc/blog3/mon-ca-ba-sa-vi-cb.html"
pasagius_recipes = Website(pasagius_url)

print(pasagius_recipes.title)
print(pasagius_recipes.text)
````

````python
# Here we have two type of prompt for the models

# System prompts: tell tasks and what tone should the response be.
system_prompt = "You are an amazing homecook that analyzed the recipes from a website \
and provides a short summary of different recipes, \
ignore any navigation, text and unrelated link. Respond in markdown and in English"
````

````python
# User prompts: The starter of the conversation
def user_prompt_for(website):
    return f"""You are looking at a website titled {website.title}. \
Please provide a summary of recipes from the website in markdown, \
ignore any advertisements and unrelated links. The content of website is as followed; \
{website.text}
    """
  

#NOTE: The credit belongs to Ed Donner from Udemy course
````

````python
# Check the content
user_prompt_for(pasagius_recipes)
````

### Message for API from OpenAI

The message should have the structure of different dicts with 2 keys inside a list
```
message =
[
    {"role" : "system", "content": "system prompt"}
    {"role" : "user", "content": "user prompt"}
]

````python
# Define a function to create the format
def message_for_web(website):
    # API from OpenAI expect the structure above
    return [
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": user_prompt_for(website)}
    ]
message_for_web(pasagius_recipes)
````

Put everything together

````python
def get_recipes_from(website):
    open_ai_ollama = OpenAI(base_url="http://localhost:11434/v1", api_key="ollama")

    messages = message_for_web(website)

    response_from_ollama = open_ai_ollama.chat.completions.create(
        model="llama3.2",
        messages= messages
    )
    print(response_from_ollama.choices[0].message.content)
````

````python
# Function to display the output better
def pprint(website):
    recipes = get_recipes_from(website)
    display(Markdown(recipes))
````

````python
pprint(pasagius_recipes)
````
