### DAY 1 Web Summary with Frontier Models

This is my first project working with frontier LLMs models. The goal is to create a web browser, where I fetch a URL to an models and it will response a summary. Based on the prompts (system and user prompt), the models can response in different texts with different tones.

```python
# Import necessary libraries
import os
import requests
from bs4 import BeautifulSoup
from IPython.display import display, Markdown
from openai import OpenAI
```

```python

# A simple greeting to model
open_ai_ollama = OpenAI(base_url="http://localhost:11434/v1", api_key="ollama")

my_message = [
    {
        "role": "user",
        "content": "Hello, this is my very first call to an "
    }
]

gemma2 = open_ai_ollama.chat.completions.create(
    model="gemma2:2b",
    messages=my_message
)
print(gemma2.choices[0].message.content)
```

### A simple web fetching function

```python

# A class to represent a Webpage
# Some websites need you to use proper headers when fetching them:
headers = {
 "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/117.0.0.0 Safari/537.36"
}

class Website:

    def __init__(self, url):
        """
        Create this Website object from the given url using the BeautifulSoup library
        """
        self.url = url
        response = requests.get(url, headers=headers)
        soup = BeautifulSoup(response.content, 'html.parser')
        self.title = soup.title.string if soup.title else "No title found"
        for irrelevant in soup.body(["script", "style", "img", "input"]):
            irrelevant.decompose()
        self.text = soup.body.get_text(separator="\n", strip=True)

#NOTE: The credit belongs to Ed Donner from Udemy course
```

```python
# Lets try with actual url
pasagius_url = "https://memart.vn/tin-tuc/blog3/mon-ca-ba-sa-vi-cb.html"
pasagius_recipes = Website(pasagius_url)

print(pasagius_recipes.title)
print(pasagius_recipes.text)
```
