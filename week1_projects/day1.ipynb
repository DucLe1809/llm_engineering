This is the first line of code to learn about calling API from open AI to experiment different LLMs models locally on my computer

```python
# Import necessary libraries
import os
from openai import OpenAI

```

```python
#
!ollama pull gemma2:2b
open_ai_ollama = OpenAI(base_url="http://localhost:11434/v1", api_key="ollama")

my_message = [
    {
        "role": "user",
        "content": "Hello, this is my very first call to an "
    }
]

gemma2 = open_ai_ollama.chat.completions.create(
    model="gemma2:2b",
    messages=my_message
)
print(gemma2.choices[0].message.content)
```
